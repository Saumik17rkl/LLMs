```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-2b-it", use_auth_token=True)

# Load the model
model = AutoModelForCausalLM.from_pretrained("google/gemma-2-2b-it", use_auth_token=True)

def predict(query):
    # Preprocess the input query using the tokenizer
    inputs = tokenizer(query, return_tensors="pt")

    # Get the model's prediction
    outputs = model.generate(**inputs)

    # Decode the output to text using the tokenizer
    result = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return result

# Create the Gradio interface
iface = gr.Interface(fn=predict, inputs="text", outputs="text")

# Launch the app
iface.launch()
```
